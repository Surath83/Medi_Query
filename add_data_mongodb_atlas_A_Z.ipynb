```python
import pandas as pd

# Load dataset
df = pd.read_csv("./Extensive_A_Z_medicines_dataset_of_India.csv")

# Columns to remove
cols_to_drop = ["Is_discontinued"]

# Drop the unwanted columns
df_cleaned = df.drop(columns=cols_to_drop, errors="ignore")

# Drop exact duplicate rows (like the Allegra duplicate in your sample)
df_cleaned = df_cleaned.drop_duplicates()

# Save cleaned dataset
df_cleaned.to_csv("medicines_cleaned.csv", index=False)

print("✅ Cleaning done! Saved as medicines_cleaned.csv")
print(df_cleaned.head(1))
```

```python
df = pd.read_csv("medicines_cleaned.csv")

# Replace empty strings, whitespace-only strings, and NaN with None
df = df.applymap(lambda x: None if (pd.isna(x) or str(x).strip() == "") else x)

# Save the modified dataset
df.to_csv("medicines_cleaned_filled.csv", index=False)

print("✅ Empty cells replaced with None. Saved as medicines_cleaned_filled.csv")
df.head(5)
```
